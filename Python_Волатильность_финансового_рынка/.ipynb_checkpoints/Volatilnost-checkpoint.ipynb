```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
X = pd.read_csv('stationary.csv')
X = X.to_numpy()
plt.plot(X)
```

Среднее




Дисперсия


```python
n = len(X)
mu = np.mean(X) #среднее арифметическое
S = sum((X[i] - mu)**2 for i in range(n))/(n-1)#дисперсия
x_norm=np.zeros(n)
for i in range(n):
    x_norm[i] = (X[i] - mu)/ np.sqrt(S) #центрирование и нормирование 
plt.plot(x_norm)
```

Автоковариация и автокорреляция


```python
def autocov(X,k):
    n=len(X)
    c=np.empty(k)
    for i in range(k):
        c[i]=sum(X[j]*X[j+i] for j in range(n-i))/(n-i)
    return c

def autocorr(c,k):
    r=np.empty(k)
    for i in range(1,k):
        r[i]=c[i]/c[0]
    return r

k=11
c=autocov(X,k)
r=autocorr(c,k)
```

Частная автокорреляция 


```python
k=11
def matr_chisl(k, r):
    mas = [[1 for i in range(k)] for j in range(k)]
    for i in range(k):
        for j in range(k-1):
            if i > j:
                mas[i][j] = r[i-j]
            elif j > i:
                mas[i][j] = r[j-i]
            elif i == j:
                mas[i][j] = 1

        mas[i][k-1] = r[i+1]
    return mas

def matr_znam(k, r):
    mas = [[1 for i in range(k)] for j in range(k)]
    for i in range(k):
        for j in range(k):
            if i > j:
                mas[i][j] = r[i-j]
            elif j > i:
                mas[i][j] = r[j-i]
            elif i == j:
                mas[i][j] = 1
    return mas

psi=np.ones(k)
for i in range(1, k):
    matr_ch = matr_chisl(i, r)
    matr_zn = matr_znam(i, r)
    det1=np.linalg.det(matr_ch)
    det2=np.linalg.det(matr_zn)
    psi[i] = det1/det2
```

```python
r=np.delete(r,0)

fig, ax = plt.subplots()
ax.bar(np.arange(1,k), r)
ax.set_title("АКФ")

psi=np.delete(psi,0)
    
fig, ax = plt.subplots()
ax.bar(np.arange(1,k), psi)
ax.set_title("ЧАКФ")
```

Уравнение для AR(1)


```python
phi=r[0]
S_eps=c[0]-phi*c[1]
print("Phi",phi)
```

```python
def prediction(steps, X, phi):
    X_pred = np.concatenate((X, np.zeros(steps)), axis = None)
    n = len(X)
    for i in range(steps):
        X_pred[n+i] = phi* X_pred[n+i-1]
    return X_pred

steps=5
X_pred=prediction(steps,X,phi)

for i in range(1,steps+2):
    print(X_pred[-i])

v=X_pred[-(steps+1):]
z=np.arange(500,506,1)
```
